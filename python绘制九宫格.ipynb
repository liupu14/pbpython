### 导入必要的python库

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 设置绘图背景以及显示

```python
sns.set_style('darkgrid')
%matplotlib inline
```

### 创建数据集

```python
np.random.RandomState(seed=1314) # 设置随机数种子，以便能够复现
Dta = {'IndexA':np.random.rand(50), 
       'IndexB':np.random.rand(50), 
       'Citys':['City' + str(ii) for ii in range(1,51)], 
       'Class':["ClassA","ClassB","ClassC","ClassD","ClassE"] * 10}
Data = pd.DataFrame(Dta)
Data.head()
```

### 绘制图形

#### 散点图绘制

```python
sns.relplot(x="IndexA",y="IndexB",data=Data)
```

#### 绘制分类散点图

```python
sns.relplot(x="IndexA",y="IndexB",data=Data,hue="Class")
```

#### 给散点图增加数据标签

```python
sns.relplot(x="IndexA",y="IndexB",hue="Class",data=Data)
for ii,jj,zz in zip(Data.IndexA,Data.IndexB,Data.Citys):
    plt.text(ii,jj,zz)
```

#### 绘制九宫格散点图

```python
sns.relplot(x="IndexA",y="IndexB",hue="Class",data=Data)
for ii,jj,zz in zip(Data.IndexA,Data.IndexB,Data.Citys):
    plt.text(ii,jj,zz)
plt.hlines(y=np.percentile(Data.IndexB,33.3),xmin=np.min(Data.IndexA),xmax=np.max(Data.IndexA),colors="red")
plt.hlines(y=np.percentile(Data.IndexB,66.7),xmin=np.min(Data.IndexA),xmax=np.max(Data.IndexA),colors="red")
plt.hlines(x=np.percentile(Data.IndexA,33.3),ymin=np.min(Data.IndexB),ymax=np.max(Data.IndexB),colors="red")
plt.hlines(x=np.percentile(Data.IndexA,33.3),ymin=np.min(Data.IndexB),ymax=np.max(Data.IndexB),colors="red")
```
